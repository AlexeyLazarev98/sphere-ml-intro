## Аналитическое решение для SIS

Вспомним как выглядит система диффуров для модели SIS:<br\>
\begin{equation}
   \begin{cases}
   \cfrac{ds_i(t)}{dt} = -\beta s_i(t)\sum\limits_j A_{ij}x_j(t) + \gamma x_i(t)\\
   \cfrac{dx_i(t)}{dt} = \beta s_i(t)\sum\limits_j A_{ij}x_j(t) - \gamma x_i(t)
  \end{cases}
  \\
  x_i(t) + s_i(t) = 1
\end{equation}
где $x_i(t)$ и $s_i(t)$ задают состояние для вершины $v_i$ быть инфицированной или восприимчивой к инфицированию.

Попробуем аналитически решить эту систему для некоторого графа

```python
from numpy.linalg import eig
from scipy.integrate import odeint
```

```python
# Генерируем граф
n = 1000
n_infected = 5
G = ig.Graph.Barabasi(n, 15)

# Достаем матрицу смежности (не самый лучший способ)
A = np.array(G.get_adjacency().data)

# Задаем коэффициенты модели
beta, gamma = 0.03, 0.2

# Задаем временной интервал
t = np.arange(0,5,0.05)

# Задаем начальное состояние
idx = np.random.choice(range(n), n_infected)
i0 = np.zeros((n,))
i0[idx] = 1
z0 = np.r_[1-i0,i0]

# Пишем систему
def sis(z, t, A, n, beta, gamma):
    return np.concatenate((
        -beta * z[0:n] * A.dot(z[n:2*n]) + gamma * z[n:2*n],
         beta * z[0:n] * A.dot(z[n:2*n]) - gamma * z[n:2*n]))
        

# Решаем
z = odeint(sis, z0, t, (A, n, beta, gamma))
```

```python
z.shape
```

```python
nId = 6
s = z[:,0:n]
x = z[:,n:2*n]


fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(s[:, nId], color = 'blue')
ax.plot(x[:, nId], color = 'red')
ax.set_xlabel('$t$')
ax.set_ylabel('prob')
ax.set_title('Probability for node %d' % nId, fontsize = 15)
```

* if $\lambda_1 < \frac{\gamma}{\beta}$ - Ослабление инфекции <br/>

   \cfrac{dx_i(t)}{dt} = \beta s_i(t)\sum\limits_j A_{ij} x_j(t) - \gamma x_i(t)\\
   \cfrac{dr_i(t)}{dt} = \gamma x_i(t)
